## 2. Numerical Solutions to ODEs

```python
## 2a

# defining numerical method for solving differential equations
def euler(dxdt, x0, tstop, dt):
    # func is the first derivative of x(t) wrt t
    # x0 is the value of x(t) at t = 0
    
    #preallocating list to hold iterations of ODE
    x_i1 = []

    for i in np.arange((int(tstop/dt))):
        
        #defining list containing all iterations of x_i+1
        x_i1 += [x0 + dxdt(x0,dt)*dt]
        
        #updating x0 with xi_1 of current iteration
        x0 = x_i1[i]
        
    return x_i1
```

```python
## 2b

#defining the ordinary differntial equations in part 1
def dadt(x,t):
    return -x

def dbdt(x,t):
    return x**-1

def dcdt(x,t):
    return 1 - x

#the derivative to d, the differential equation we'll solve
def logistic(x,t):
    return x*(1-x)

#defining numerical solutions to each equation
nsa = euler(dadt, 1, 5, .01)
nsb = euler(dbdt, 1, 5, .01)
nsc = euler(dcdt, 0, 5, .01)
nsd = euler(logistic, .5, 5, .01)

#defining analytical solutions:
def asa(t):
    return (np.e**-t).reshape(len(t),1)
def asb(t):
    return np.sqrt(2*t + 1).reshape(len(t),1)
def asc(t):
    return (1 - np.e**-t).reshape(len(t),1)
```

```python
t = np.linspace(0,5,500)

plt.subplot(2,2,1)
plt.plot(t, nsa, label = 'numerical')
plt.plot(t, asa(t), label = 'analytical')
plt.xlabel('t')
plt.ylabel('a(t)')
plt.legend()

plt.subplot(2,2,2)
plt.plot(t, nsb, label = 'numerical')
plt.plot(t, asb(t), label = 'analytical')
plt.xlabel('t')
plt.ylabel('b(t)')
plt.legend()

plt.subplot(2,2,3)
plt.plot(t, nsc, label = 'numerical')
plt.plot(t, asc(t), label = 'analytical')
plt.xlabel('t')
plt.ylabel('c(t)')
plt.legend()

plt.subplot(2,2,4)
plt.plot(t, nsd, label = 'numerical')
plt.xlabel('t')
plt.ylabel('d(t)')
plt.legend()

plt.tight_layout()
plt.show()
```

2b. The numerical solution serves as a good approximator but the numerical solution gives us a better representation of x, because it it not calculated in steps. Rather, it is calculated using continuous integration. The smaller the step size for the numerical solution, the smoother and more accurate the curve will be.

```python
from scipy import integrate,arange
```

```python
## 2c
    
#setting hyperparameters
t = arange(0,5,.01)
x_0 = 0

#using scipy method for 1c ODE solution
sci_c = integrate.odeint(dcdt,x_0,t)
```

```python
#defining squared error
def se(y,yt):
    return (y - yt)**2

# reshaping the euler solutin to 1c in preparation for pass to squared
# error function
def plot_euler_c(c):
    return np.asarray(c).reshape(len(c),1)

#Plotting the squared error for different dt

# dt = .01
t = arange(0,5,.01)
xc = euler(dcdt, 0, 5, .01)
sci_c = integrate.odeint(dcdt,x_0,t)

plt.figure()
plt.plot(t, se(plot_euler_c(xc), asc(t)), label = 'Euler C')
plt.plot(t, se(sci_c, asc(t)), label = 'Scipy C')
plt.legend()
plt.xlabel('t')
plt.ylabel('Squared Error')
plt.title('Numerical Solution Error, dt = .01')

# dt = .1
t = arange(0,5,.1)
xc = euler(dcdt, 0, 5, .1)
sci_c = integrate.odeint(dcdt,x_0,t)

plt.figure()
plt.plot(t, se(plot_euler_c(xc), asc(t)), label = 'Euler C')
plt.plot(t, se(sci_c, asc(t)), label = 'Scipy C')
plt.legend()
plt.xlabel('t')
plt.ylabel('Squared Error')
plt.title('Numerical Solution Error, dt = .1')

# dt = 1
t = arange(0,5,1)
xc = euler(dcdt, 0, 5, 1)
sci_c = integrate.odeint(dcdt,x_0,t)

plt.figure()
plt.plot(t, se(plot_euler_c(xc), asc(t)), label = 'Euler')
plt.plot(t, se(sci_c, asc(t)), label = 'Scipy')
plt.legend()
plt.xlabel('t')
plt.ylabel('Squared Error')
plt.title('Numerical Solution Error, dt = 1')

plt.tight_layout()
plt.show()
```

The scipy integrate.odeint function does a much better job of approximating the analytical solution that our implemented euler method. As the steps become larger, the euler numerical solution has exponentially higher error.

## 3. Passive Membrane

```python
## 3a

#Setting hyperparameters, defining in seconds, volts, amperes, and ohms
Rm = 10**7 #ohms
it = I0 = 1e-9 #amperes
tm = 10e-3 #seconds
Em = -80e-3 #volts

def dVdt(Vt,dt):
    return (-Vt + Em + Rm*it)/tm

t = arange(0,.08,1e-3)
# voltage_change = euler(dVdt,Em,.08,1e-3) #Euler's method
voltage_change = integrate.odeint(dVdt,Em,t) #Scipy method


plt.plot(t, voltage_change)
plt.xlabel('time (in seconds)')
plt.ylabel('Voltage (in volts)')
plt.title('Voltage Change over Time')
plt.show()
```

```python
## 3b

#defining sinusoidal time-dependent current
def I(dt,v): 
    return I0*np.sin(2*np.pi*v*dt)

#creating array with all v's
v = np.array((1,10,30))

for i in np.arange(len(v)):

    #defining passive membrane voltage change equation
    def dVdt(Vt,dt):
        return (-Vt + Em + Rm*I(dt,v[i]))/tm
    
    #setting time step, ODE solutions resolution of 1000 time points
    t = np.linspace(0,10/v[i],1000,endpoint=False)

    #finding numerical solution to ODE with scipy
    voltage_change = integrate.odeint(dVdt,Em,t)

    # plotting curves of voltage vs. time
    plt.figure()
    plt.plot(t, voltage_change, label = 'Voltage change')
    plt.plot(t, Em+Rm*I(t,v[i]), label = 'Rescaled Voltage')
    plt.xlabel('time (in seconds)')
    plt.ylabel('votlage (in volts)')
    plt.title('v = %d' % v[i])
    plt.legend()
    plt.show()
```

```python
#3c

#defining frequency 1 - 100 Hz
v = np.arange(1,101,1)

# preallcoating array to hold information about maximum of last simulated
# period, allowing us to calculage phase lag
volt_hist = np.zeros((len(v),1000,1))
rescaled_V = np.zeros((len(v),1000,1))

#Preallocating arrays for phase lag and amplitude
philag = np.zeros((len(v-1),1))
amp = np.zeros((len(v-1),1))

#setting time step, ODE solutions resolution of 1000 time points
t = np.linspace(0,1,1000,endpoint=False)    

#looping over 100 frequency values
for i in np.arange(len(v)): #TESTING: REPLACE WITH v LATER
        
    #defining passive membrane voltage change equation for current frequency
    def dVdt(Vt,dt):
        return (-Vt + Em + Rm*I(dt,v[i]))/tm
    
    #defining rescaled voltage, the driving oscillation
#     rescaled_V[i] = (Em+Rm*I(t,v[i])).reshape(1000,1)

    #or it is the sinusoidal time-dependent current alone?
    rescaled_V[i] = np.sin(2*t*np.pi*v[i]).reshape(1000,1)
    
    #defining numerical solution to passive membrane voltage
    volt_hist[i] = integrate.odeint(dVdt,Em,t)

    #defining size of one period
    period_size = int(1000/v[i])
    
    # creating index for calculation of amp and philag
    p = -period_size
    
    #calcuting phase shift from the final wave period of current frequency
    # and the driving oscillation
    philag[i] = (np.argmax(volt_hist[i][p:999]) - \
         np.argmax(rescaled_V[i][p:999]))/period_size

    #taking the max - min value of the final period for current frequency
    amp[i] = np.max(volt_hist[i][p:999]) - np.min(volt_hist[i][p:999])
```

```python
#defining analytical solution for amplitude and phase lag
def A(v):
    return I0*Rm*np.sqrt(1/(1+(np.pi*tm*v)**2)) 

def dphi(v):
    return np.arctan(2*np.pi*v*tm)
    

# plotting current frequency v vs. amplitude A & analytical solution
plt.figure()
plt.plot(v, A(v), label = 'analytical')
plt.plot(v, amp, label = 'our solution')
plt.xlabel('current frequency')
plt.ylabel('amplitude')
plt.title('Dependency of Amplitude on Frequency')
plt.legend()


# plotting current frequency v vs. phase diferrences dphi & 
# analytical solution
plt.figure()
plt.plot(v,dphi(v),label = 'analytical')
plt.plot(v,philag,label = 'our solution')
plt.xlabel('current frequency')
plt.ylabel('phase lag')
plt.title('Dependency of Phase Lag on Frequency')
plt.legend()
plt.show()
```
