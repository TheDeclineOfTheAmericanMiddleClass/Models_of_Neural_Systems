**1. Visual Receptive Fields**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1a
x0 = 5  #setting size of visual field (5+5)^2
y0 = 5

#setting delta_x and delta_y
dx = dy = .01
i = j = np.arange(2*x0/dx)

x = dx*i-x0
y = dy*j-y0

#creating holding container for two arrays x and y
visField = np.meshgrid(x,y)
```

```python
#1b
Phi = 0
A = 1
K = np.pi
theta = 0/360

def grating(X,Y,THETA,kay,phi):
    s = A*np.cos(K*X*np.cos(THETA) + kay*Y*np.sin(THETA) - phi)
    return s

grid = grating(visField[0],visField[1],theta,K,Phi)
plotty = plt.imshow(grid)

plt.xlabel('x-direction')
plt.ylabel('y-direction')
plt.title('Sinusoidal Grating')
plt.colorbar(plotty)
```

```python
c_theta = np.linspace(0,360,6)
print(c_theta)

c_k = np.random.normal(1.7,1,6)
print(c_k)

c_theta[1].round(decimals=1)
```

```python
#c
c_theta = np.linspace(0,360,6)
print(c_theta)

c_k = np.random.normal(1.7,1,6)
print(c_k)

for k in range(6): #plotting 5 figures of varying theta/k
    grid = grating(visField[0],visField[1],c_theta[k],c_k[k],Phi)
    fig, ax = plt.subplots()
    ax = plt.imshow(grid)
    plt.title(" = %f, k = %f" % (np.around(c_theta[k],decimals=2), \
              np.around(c_k[k],decimals=2)))
    plt.colorbar(ax)
```

**2. A Model of Receptive Fields**

```python
#a 
phase = 0
sX = 2/np.pi
sY = 1/np.pi
k = np.pi*2

def Gabor(x,y):
    D = 1/(2*np.pi*sX*sY)*np.exp(-(x**2/(2*sX) + y**2/(2*sY))) \
                    *np.cos(k*x-phase)
    return D

RF = Gabor(visField[0],visField[1])
plt.imshow(RF)
plt.show()
```

#b 

The optimal stimuli for a neuron with this RF are those where K is preferred spatial frequency and (big) phi is preferred spatial phase. This RF characterizes a simple cell becuase its RF can be plotted with the Gabor function. Thus, it looks for edges at certain location  and of certain orientations, which are very low level visual elements. A neuron like this could be found in the primary visual cortex, V1. 

```python
#c
#Calculating dot product of cell's receptive field and sinusoidal grating

#calculating response to visual stimulus s
rs = dx*dy*np.multiply(grid,RF).sum(axis=0).sum(axis=0)

#calculating response to a uniform field
A = np.ones((1000,1000))*A
ru = dx*dy*np.multiply(A,RF).sum(axis=0).sum(axis=0)
```

```python
print('The response of the neuron to visual stimuli s was',rs)
print('The response of the neuron to a uniform field was',ru)
```

The difference in response of the neuron to the uniform field vs. the visual stimulus s was due to the presence of the preference edges of k preferred spatial frequency and phi preferred spatial phase in the field of each. The uniform field had no edges so the neuron's response was close to 0. On the other hand, even though the spatial frequency and phase weren't optimal for the neuron, it still had a response that was 1/2 of the response it would have if it were presented a stimulus with its preferred attributes.

**3. Tuning Curve**

```python
#3a
theta = np.linspace(0,np.pi) #50 different angles for theta
stim_ori = np.zeros((len(theta),1000,1000)) #preallocating matrix for thetas
rs = np.zeros((theta.shape))

for i in range(len(theta)):
    #creating gratings of 50 different thetas
    stim_ori[i] = grating(visField[0],visField[1],theta[i],0,0)
    rs[i] = dx*dy*np.multiply(stim_ori[i],RF).sum(axis=0).sum(axis=0)
    
plt.plot(theta,rs)
plt.xlabel('Theta (in radians) \n k = 0, phi = 0')
plt.ylabel('Response of Neuron')
plt.show()
```

```python
#3b
k = np.linspace(0,np.pi) #50 different angles for k
stim_freq = np.zeros((len(k),1000,1000)) #preallocating matrix for k's
rs = np.zeros((k.shape))

for i in range(len(k)):
    #creating gratings of 50 different thetas
    stim_freq[i] = grating(visField[0],visField[1],1,k[i],0)
    rs[i] = dx*dy*np.multiply(stim_freq[i],RF).sum(axis=0).sum(axis=0)
    
plt.plot(K/k,rs)
plt.xlabel('Stimulus Frequency/Neuron Preferred Frequency \n theta = 1, phi = 0')
plt.ylabel('Response of Neuron')
plt.show()
```

```python
#3b
Phi = np.linspace(0,np.pi*2) #50 different angles for phi
stim_pha = np.zeros((len(Phi),1000,1000)) #preallocating matrix for phis
rs = np.zeros((Phi.shape))

for i in range(len(Phi)):
    #creating gratings of 50 different thetas
    stim_pha[i] = grating(visField[0],visField[1],1,0,Phi[i])
    rs[i] = dx*dy*np.multiply(stim_pha[i],RF).sum(axis=0).sum(axis=0)
    
plt.plot(Phi,rs)
plt.xlabel('Stimulus Spatial Phase \n theta = 1, k = 0')
plt.ylabel('Response of Neuron')
plt.show()
```

The receptive field prefers stimuli in phase 0 degrees, at orientation 0 degrees, and a spatial frequency as a high as possible (although it does seem to hit an upper limit at 8*pi).

**4. Bonus Exercise: Image Filtering**

```python
#4a

#reads in RGB image
coveC = plt.imread('cove.jpg')

#function for grayscale conversion
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])

#Converts image to grayscale   
cove = rgb2gray(coveC)    
plt.imshow(cove, cmap = plt.get_cmap('gray'))
plt.show()
```

```python
#4b
import scipy.signal as sci

#setting preferred stimulus parameters
phase = np.pi/2
sX = np.pi*1/8
sY = np.pi*1/8
k = np.pi*8

#setting range of receptive field
x0 = y0 = 5
i = j = np.arange(2*x0/dx)

#calculacing neuron's Gabor
dx = dy = .05
x = dx*i-x0
y = dy*i-y0
field = np.meshgrid(x,y)

def Gabor(x,y):
    D = 1/(2*np.pi*sX*sY)*np.exp(-(x**2/(2*sX) + y**2/(2*sY))) \
                    *np.cos(k*x-phase)
    return D


Ds = Gabor(field[0],field[1])
conv = sci.fftconvolve(cove,Ds,'same')
```

```python
plt.imshow(conv)
```

The gabor filter is detecting the presence of edges in this picture as its preferred orientation, spatial frequency, and phase frequency. It will "fire" more at places in the picture where there are edges who characteristics are closer to these preferred parameters. It would be good for identifying contours in images. In self-driving cars, for example, it could be used to identify the painted streak on a highway, and be part of a process to ensure the car is driving on the road.
